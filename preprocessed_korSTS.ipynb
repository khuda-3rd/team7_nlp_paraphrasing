### Train, Test 합치고 유사 문장만 저장
labels 값이  점 이상이어야 유사한 문장임을 확인
- 그것만 남기고 저장

```python
train
```

```python
train = train[train['labels']>=4.5]
train
```

```python
test
```

```python
test = test[test['labels'] >= 4.5]
test
```

- 우선 train, test 위아래로 합치기

```python
# 위/아래로 합치기 - 행 기준
df = pd.concat([train, test], axis = 0)
df
```

```python
df.drop(columns = 'labels', inplace = True)
```
