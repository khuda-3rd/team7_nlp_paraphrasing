```python
import pandas as pd
import json

# JSON 파일 불러오기
filepath = '/content/drive/MyDrive/NLP프로젝트/박민아/국립국어원 유사 문장 말뭉치(버전 1.0)/NIKL_PC.json'
with open(filepath, 'r', encoding='utf-8') as file:
    data = json.load(file)

# DataFrame 생성
df = pd.DataFrame(columns=['sentence_id', 'form', 'generation'])

# JSON 데이터를 DataFrame에 추가
rows = []
for item in data['data']:
    sentence_id = item['sentence_id']
    paraphrases = item['paraphrases']
    for paraphrase in paraphrases:
        form = paraphrase['form']
        generation = paraphrase['generation']
        rows.append({'sentence_id': sentence_id, 'form': form, 'generation': generation})

df = pd.concat([df, pd.DataFrame(rows)], ignore_index=True)

# DataFrame 출력
print(df)
```

```python
# 기계가 만든 데이터 삭제
filtered_df = df[df['generation'] != 'machine']
```

```python
# 데이터프레임의 공백 칸 검사
def check_dataframe_empty(df):
    empty_cells = df.isnull().sum().sum()
    if empty_cells > 0:
        print("데이터프레임에 공백 칸이 있습니다.")
    else:
        print("데이터프레임에 공백 칸이 없습니다.")

# 데이터프레임에 적용
check_dataframe_empty(filtered_df)
```

```python
import pandas as pd

# 기존 데이터프레임에서 sentence_id가 같은 행의 form을 합쳐서 새로운 데이터프레임 생성
def combine_forms(df):
    new_rows = []
    grouped = df.groupby('sentence_id')

    for sentence_id, group in grouped:
        forms = group['form'].tolist()
        new_row = {'sentence_id': sentence_id}
        for i, form in enumerate(forms, start=1):
            new_row[f'form{i}'] = form
        new_rows.append(new_row)

    new_df = pd.concat([pd.DataFrame([new_rows[i]]) for i in range(len(new_rows))], ignore_index=True)
    return new_df

# 새로운 데이터프레임 생성
new_df = combine_forms(filtered_df)

# 새로운 데이터프레임 출력
print(new_df)
```

```python
df = df.drop(["generation"], axis = 1)
```

```python
# 유사한 문장의 개수 확인 
df[df["sentence_id"]=='NWRW1800000021.1.2.1'].shape[0]
```

```python
# sentence_id를 순차적으로 변경
df['sentence_id'] = df.groupby('sentence_id').ngroup()

# 결과 확인
print(df)
```

```python
df.shape[0]
```

```python
sentence_pairs = []
for sentence_idx in range(df["sentence_id"].nunique()):
    df_sliced = df[df["sentence_id"] == sentence_idx]
    for i in range(1, df_sliced.shape[0]):
        for j in range(i):
            sentence_pairs.append([df_sliced["form"].values[i], df_sliced["form"].values[j]])
```

```python
sentence_pairs
```

```python
paired_df = pd.DataFrame(sentence_pairs, columns = ["texts", "pairs"])
```
